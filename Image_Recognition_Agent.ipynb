```python
def split_train_test(folder_path, train_folder, test_folder, test_percentage):

    files = os.listdir(folder_path)


    random.shuffle(files)


    num_test = int(len(files) * test_percentage)


    train_files = files[num_test:]
    test_files = files[:num_test]


    for filename in train_files:
        src_path = os.path.join(folder_path, filename)
        dest_path = os.path.join(train_folder, filename)
        copyfile(src_path, dest_path)


    for filename in test_files:
        src_path = os.path.join(folder_path, filename)
        dest_path = os.path.join(test_folder, filename)
        copyfile(src_path, dest_path)


rock_0_grayscale_folder = '/content/drive/MyDrive/image_dataset/rock_0_grayscale'
scissors_1_grayscale_folder = '/content/drive/MyDrive/image_dataset/scissors_1_grayscale'
paper_2_grayscale_folder = '/content/drive/MyDrive/image_dataset/paper_2_grayscale'


train_folder = '/content/drive/MyDrive/image_dataset/train'
test_folder = '/content/drive/MyDrive/image_dataset/test'


if not os.path.exists(train_folder):
    os.makedirs(train_folder)

if not os.path.exists(test_folder):
    os.makedirs(test_folder)


split_train_test(rock_0_grayscale_folder, train_folder, test_folder, 0.2)


split_train_test(scissors_1_grayscale_folder, train_folder, test_folder, 0.2)


split_train_test(paper_2_grayscale_folder, train_folder, test_folder, 0.2)
```

```python
rock_0_grayscale_folder = '/content/drive/MyDrive/image_dataset/rock_0_grayscale'
scissors_1_grayscale_folder = '/content/drive/MyDrive/image_dataset/scissors_1_grayscale'
paper_2_grayscale_folder = '/content/drive/MyDrive/image_dataset/paper_2_grayscale'

def create_dataframe(folder, label):
    file_paths = [os.path.join(folder, file) for file in os.listdir(folder) if file.endswith(('.jpg', '.png'))]
    df = pd.DataFrame({'file_path': file_paths, 'label': label})
    return df

rock_df = create_dataframe(rock_0_grayscale_folder, 'class_0')
scissors_df = create_dataframe(scissors_1_grayscale_folder, 'class_1')
paper_df = create_dataframe(paper_2_grayscale_folder, 'class_2')

all_images_df = pd.concat([rock_df, scissors_df, paper_df], ignore_index=True)

img_width, img_height = 300, 200
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_generator = train_datagen.flow_from_dataframe(
    dataframe=all_images_df,
    x_col='file_path',
    y_col='label',
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)


validation_generator = train_datagen.flow_from_dataframe(
    dataframe=all_images_df,
    x_col='file_path',
    y_col='label',
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

print(f"Found {train_generator.samples} training images belonging to {len(train_generator.class_indices)} classes.")
print(f"Found {validation_generator.samples} validation images belonging to {len(validation_generator.class_indices)} classes.")

model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(img_width, img_height, 3), kernel_regularizer=regularizers.l2(0.001)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(128, (3, 3), activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)))
model.add(Dropout(0.5))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer=Adam(lr=0.01), loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=80,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)

model.save('/content/sample_data/model.h5')
```

```python
test_folder = '/content/drive/MyDrive/image_dataset/final_images'

all_test_image_filenames = [filename for filename in os.listdir(test_folder) if filename.endswith(('.jpg', '.png'))]

test_image_filenames = random.sample(all_test_image_filenames, 20)

saved_model_path = '/content/sample_data/model.h5'
loaded_model = load_model(saved_model_path)

stack = 10
wins = 0
loses = 0
draws = 0

def choice(opp):
  if opp==0:
    return 2
  elif opp==1:
    return 0
  else:
    return 1

def bet(opp,me):#no pool

  global stack
  global wins
  global loses
  global draws

  if opp==0 and me==1:
    stack = stack - 1
    loses = loses + 1
  elif opp==0 and me==2:
    stack = stack + 1
    wins = wins + 1
  elif opp==1 and me==0:
    stack = stack + 1
    wins = wins + 1
  elif opp==1 and me==2:
    stack = stack - 1
    loses = loses + 1
  elif opp==2 and me==0:
    stack = stack - 1
    loses = loses + 1
  elif opp==2 and me==1:
    stack = stack + 1
    wins = wins + 1
  else:
    pass
    draws = draws + 1
  print(stack)
  print(opp,me)



def preprocess_image(image_path, target_size=(300, 200)):
    img = image.load_img(image_path, target_size=target_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)
    img_array /= 255.0
    return img_array

def predict_image(model, image_path):
    processed_image = preprocess_image(image_path)
    predictions = model.predict(processed_image)
    class_index = np.argmax(predictions[0])
    return class_index

for image_filename in test_image_filenames:
    image_path = os.path.join(test_folder, image_filename)


    first_char = image_filename[0]
    print(f"Image Name: {image_filename}, First Character: {first_char}")


    predicted_class = predict_image(loaded_model, image_path)
    print(f"Model Prediction: Class {predicted_class}")
    print("-------------------")
    bet(int(first_char),choice(predicted_class))
print("-------------------")
print("The agent finishes the game with a total of",stack,"coins")
print("Also with",wins,"wins",loses,"loses and",draws,"draws")
```
